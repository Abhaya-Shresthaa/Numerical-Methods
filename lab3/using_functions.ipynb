Least Square Method

i. Linear cureve fitting. y = a +bx
ii. Exponential curve fitting y = ae^bx
using cramer's rule

```python
import numpy as np 
import sys
import matplotlib.pyplot as plt
```

```python
def xAndyValues():
    # x = np.array([50, 70, 100,120])
    # y = np.array([12,15,21,25])
    
    #for exponential
    
    x = np.array([2, 4, 6, 8, 10, 12])
    y = np.array([3, 5,  10, 18, 48, 90])
    return x, y
```

```python
def curve(whichOne):
    x , y = xAndyValues()
    n = len(x)
    Ex = np.sum(x)
    #Ey = np.sum(y)
    Ex2 = np.sum(x**2)
    #Exy = np.sum(x*y)
    if whichOne == 'linear':
        Exy = np.sum(x*y)
        Ey = np.sum(y)
    elif whichOne == 'exponential':
        Ey = np.sum(np.log(y))
        Exy = np.sum(x*(np.log(y)))
        
    D = n* Ex2 - Ex * Ex
    Da = Ey * Ex2 - Exy *Ex
    Db = n * Exy - Ex * Ey

    try: 
        if whichOne == 'linear':
            a = Da / D
        elif whichOne == 'exponential':
            a = np.exp(Da / D)
        b = Db / D
    except ZeroDivisionError:
        sys.exit("Determinant is 0")
        
    print(a, b)
    return a, b
```

```python
def valsForPlotting(whichOne):
    a, b =curve(whichOne)
    if whichOne == 'linear':
        x_line = np.linspace(0,120,400)
        y_line = a + b * x_line
    elif whichOne == 'exponential':
        x_line = np.linspace(2,12,400)
        y_line = a * np.exp(b*x_line) 
    return x_line, y_line, a, b    
```

```python
def plot(whichOne):
    x_line, y_line, a, b = valsForPlotting(whichOne)
    x, y = xAndyValues()
    if whichOne == 'linear':
        plt.plot(x_line, y_line, label = f'y = {b}x + {a}')
    elif whichOne == 'exponential':
        plt.plot(x_line, y_line, label = f'y ={a}*e^{b}x')       
    plt.scatter(x, y ,color = 'blue', label = 'Data Points', zorder = 5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
plot('linear')
```

```python
plot('exponential')
```
