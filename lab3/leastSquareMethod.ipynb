Least Square Method

i. Linear cureve fitting. y = a +bx
ii. Exponential curve fitting y = ae^bx
using cramer's rule

```python
import numpy as np 
import sys
import matplotlib.pyplot as plt
```

```python
x = np.array([50, 70, 100,120])
y = np.array([12,15,21,25])
n = len(x)
Ex = np.sum(x)
Ey = np.sum(y)
Ex2 = np.sum(x**2)
Exy = np.sum(x*y)
print(n , Ey, Ex, Exy, Ex2)
```

```python
D = n* Ex2 - Ex * Ex
Da = Ey * Ex2 - Exy *Ex
Db = n * Exy - Ex * Ey

try: 
    a = Da / D
    b = Db / D
except ZeroDivisionError:
    sys.exit("Determinant is 0")
    
print(a, b)
    
```

```python
x_line = np.linspace(0,120,400)
y_line = a + b * x_line
```

```python
plt.plot(x_line, y_line, label = f'y = {b}x + {a}')
plt.scatter(x, y ,color = 'blue', label = 'Data Points', zorder = 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

for exponential curve

```python
x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([3, 5,  10, 18, 48, 90])
n = len(x)
Ex = np.sum(x)
Ey = np.sum(np.log(y))
Ex2 = np.sum(x**2)
Exy = np.sum(x*(np.log(y)))
print(n , Ey, Ex, Exy, Ex2)
```

```python
D = n* Ex2 - Ex * Ex
Da = Ey * Ex2 - Exy *Ex
Db = n * Exy - Ex * Ey

try: 
    a = np.exp(Da / D)
    b = Db / D
except ZeroDivisionError:
    sys.exit("Determinant is 0")
    
print(a, b)
    
```

```python
x_line = np.linspace(2,12,400)
y_line = a * np.exp(b*x_line)
```

```python
plt.plot(x_line, y_line, label = f'y ={a}*e^{b}x')
plt.scatter(x, y ,color = 'blue', label = 'Data Points', zorder = 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
