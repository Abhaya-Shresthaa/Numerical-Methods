```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.5, 3.6, 7.4, 10.8, 15.3])

# Fit a 2nd-degree polynomial (quadratic)
degree = 3
coeffs = np.polyfit(x, y, degree)

# Create a polynomial function from coefficients
poly = np.poly1d(coeffs)

# Print the polynomial equation
print("Fitted Polynomial:")
print(poly)

# Generate x values for the curve
x_curve = np.linspace(min(x), max(x), 100)
y_curve = poly(x_curve)

# Plot the original data and the polynomial fit
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_curve, y_curve, color='blue', label=f'Degree {degree} Fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fitting')
plt.grid(True)
plt.show()
```
